Feature Engine Pipeline

```python
# To treat discrete variables as categorical , re-cast them as object

x_train[discrete]=x_train[discrete].astype('O')

x_test[discrete]=x_test[discrete].astype('O')
```

```python
#Import libraries required for feature engine

from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from feature_engine import missing_data_imputers as mdi
from feature_engine import categorical_encoders as ce
from feature_engine.variable_transformers import YeoJohnsonTransformer
from feature_engine.discretisers import DecisionTreeDiscretiser
from feature_engine.wrappers import SklearnTransformerWrapper
```

```python
x_train.info()
```

```python
pipe=Pipeline([
    
    # missing data imputation
    ('missing_ind', mdi.AddMissingIndicator(variables=['LotFrontage', 'MasVnrArea',  'GarageYrBlt'])),
    ('imputer_numerical', mdi.MeanMedianImputer(imputation_method ='mean',variables=['LotFrontage', 'MasVnrArea',  'GarageYrBlt'])),
    ('imputer_categorical', mdi.CategoricalVariableImputer(imputation_method='missing', variables=categorical)),
    
    #categorical encoding
    ('rare_encoding', ce.RareLabelCategoricalEncoder(tol=0.01, n_categories=3, variables=categorical + discrete)),
    ('categorical_encoding', ce.MeanCategoricalEncoder(variables=categorical + discrete)),
    
    #transforming into numerical 
    ('Yeo-Johnson', YeoJohnsonTransformer(variables=['LotFrontage','MasVnrArea', 'GarageYrBlt'])),
    
    #discretisation and encoding
    ('tree', DecisionTreeDiscretiser(cv=2, scoring='neg_mean_squared_error', regression=True, 
                                     param_grid={'max_depth': [1,2,3,4,5,6]}, 
                                     variables=['LotFrontage', 'MasVnrArea',  'GarageYrBlt'])),
    
    #feature scaling
    ('scaler', SklearnTransformerWrapper(transformer=StandardScaler()))
    
    
])
```

```python
pipe.fit(x_train, y_train)
```

```python
# Applying the transformations to train and test

x_train=pipe.transform(x_train)
x_test= pipe.transform(x_test)
```

```python
x_train
```

# Regression Models

```python
from sklearn.model_selection import GridSearchCV
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import cross_val_score
from sklearn.pipeline import make_pipeline
```

# Linear Model

```python

from sklearn.linear_model import LinearRegression

lin=Pipeline([('Scaler', MinMaxScaler()), ('Linear Reg', LinearRegression())])

lin.fit(x_train, y_train)

x_train_preds = lin.predict(x_train)
x_test_preds = lin.predict(x_test)

#Fit the pipeline
lin.fit(x_train, y_train)

#Best Score
lin_train_score = lin.score(x_train, y_train)
lin_test_score = lin.score(x_test, y_test)


print('Training set score: {:.2f}%' .format(lin_train_score*100))
print('Test set score: {:.2f}%' .format(lin_test_score*100))

```

Linear Regression Stochastic Gradient Descent

```python

from sklearn.linear_model import SGDRegressor

lin_sgd=Pipeline([('Scaler', MinMaxScaler()), ('sgd_regressor', SGDRegressor(max_iter=10000, tol=1e-6))])

param_lin_sgd={'sgd_regressor__eta0':[0.001,0.01,0.1,1,10,100]}

grid_lin_sgd = GridSearchCV(lin_sgd, param_lin_sgd, cv=4, n_jobs=-1, return_train_score = True)


#Fit the pipeline
grid_lin_sgd.fit(x_train, y_train)

#Best Score
rid_train_score = grid_lin_sgd.score(x_train, y_train)
rid_test_score = grid_lin_sgd.score(x_test, y_test)


print('Training set score: {:.2f}%' .format(rid_train_score*100))
print('Test set score: {:.2f}%' .format(rid_test_score*100))

#Best Validation score
print("Best cross-validation score: {:.2f}%".format(grid_lin_sgd.best_score_*100))

#Best parameters
print('Best parameters:',grid_lin_sgd.best_params_)
```

Ridge Regression

```python

from sklearn.linear_model import Ridge

rig_pipe=Pipeline([('scaler', MinMaxScaler()), ('ridge', Ridge())])

rig_param={'ridge__alpha': range(1,10)}

rig_grid=GridSearchCV(rig_pipe, rig_param, cv=4, return_train_score=True)

#Fit
rig_grid.fit(x_train, y_train)

#Best Score
rid_train_score = rig_grid.score(x_train, y_train)
rid_test_score = rig_grid.score(x_test, y_test)

print('Training set score: {:.2f}%' .format(rid_train_score*100))
print('Test set score: {:.2f}%' .format(rid_test_score*100))

#Best validation score
print('Best cross-validation score: {:.2f}%' .format(rig_grid.best_score_*100))

#Best parameters
print('Best parameters:',rig_grid.best_params_)
```

Tuning Multiple Models with one GridSearch

```python
model_pipe=Pipeline([('regressor', LinearRegression())])

model_params=[{'regressor':[LinearRegression()]},
              
             {'regressor':[SGDRegressor(max_iter=10000, tol=1e-6)],'regressor__eta0':[0.001,0.01,0.1,1,10,100]},
              
             {'regressor':[Ridge()], 'regressor__alpha': range(1,10)},
              
{'regressor':[Lasso(random_state=0, tol=1e-5)],'regressor__alpha':range(100,1000,100), 'regressor__max_iter':[1000]},
              
             {'regressor':[ElasticNet()],'regressor__alpha':[0.0001,0.001,0.01,0.02,1,10]},
              
             {'regressor':[KNeighborsRegressor()],'regressor__n_neighbors':range(1,25)},
              
{'regressor':[SVR()],'regressor__kernel':['rbf'], 'regressor__C':[0.001, 0.01, 0.1, 1, 10, 100], 'regressor__gamma':[0.001, 0.01, 0.1, 1, 10, 100]},
              
{'regressor':[SVR()],'regressor__kernel':['linear'],'regressor__C':[ 1, 10, 100,1000,10000]}]



model_grid=GridSearchCV(model_pipe, model_params, cv=5, return_train_score = True)

#Fit
model_grid.fit(x_train, y_train)

```

```python
#Best Score
model_train_score=model_grid.score(x_train, y_train)
model_test_score=model_grid.score(x_test, y_test)

print('Training set score: {:.2f}%' .format(model_train_score*100))
print('Test set score: {:.2f}%' .format(model_test_score*100))

#Best validation score
print('Best cross-validation score: {:.2f}%' .format(model_grid.best_score_*100))

#Best parameters
print('Best parameters:',model_grid.best_params_)
```

```python
# Predictions
x_train_pred = model_grid.predict(x_train)
x_test_pred = model_grid.predict(x_test)
```

```python
# Model performance
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from math import sqrt

print('Test mse : {:>14.2f}'.format(mean_squared_error(y_test, x_test_pred)))
print('Test rmse: {:>9.2f}'.format(sqrt(mean_squared_error(y_test, x_test_pred))))
print('Test r2  : {:>6.2f}%'.format(r2_score(y_test, x_test_pred)*100))
```
