```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats

pd.pandas.set_option('display.max_columns', None)
%matplotlib inline
```

```python
titanic.head()
```

```python
# Unique variable values

for var in titanic.columns:
    print(var,titanic[var].unique()[0:20],' \n')
```

```python
#Variable types


#mixed
mixed= ['Cabin']

#categorical
categorical=[var for var in titanic.columns if titanic[var].dtype=='O' and var not in mixed]

#Numerical:-
discrete= [var for var in titanic.columns if titanic[var].dtype !='O' and var!='Survived' and titanic[var].nunique()<10]

continuous=[var for var in titanic.columns if titanic[var].dtype !='O' and var!='Survived' and var not in discrete ]


print(f'There is {len(mixed)} mixed variables' )
print(f'There are {len(categorical)} categorical variables' )
print(f'There are {len(discrete)} discrete variables' )
print(f'There are {len(continuous)} continuous variables' )
```

```python
#Missing value

titanic.isnull().mean()
```

```python
#Cardinality (no. of different categories)

titanic[categorical+mixed].nunique()
```

```python
#To split cabin into numerical and category

titanic['Cabin_n']=titanic['Cabin'].str.extract('(\d+)') # for capturing numerical part 
# \d regular expression for digts . \d+ one or more digits
titanic['Cabin_n']=titanic['Cabin_n'].astype('float')
titanic['Cabin_c']=titanic['Cabin'].str[0] #for capturing the first letter

titanic.head()
```

```python
titanic['Title']=titanic['Name'].str.split(',').str[1].str.split('\s+').str[1]

titanic.head()
```

```python
titanic.drop(['PassengerId', 'Name', 'Ticket', 'Cabin'], axis='columns', inplace=True)
titanic.head()
```

```python
#Split the data into train and test
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(titanic.drop(['Survived'], axis=1),
                                                    titanic.Survived,
                                                    test_size=0.1,
                                                    random_state=0)
                                                    
X_train.shape, X_test.shape
```

Feature Engine


```python
from feature_engine import missing_data_imputers as mdi

from feature_engine.categorical_encoders import OneHotCategoricalEncoder
from feature_engine.categorical_encoders import RareLabelCategoricalEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from feature_engine.wrappers import SklearnTransformerWrapper
from sklearn.preprocessing import StandardScaler
```

```python
titanic_pipe=Pipeline([
    
    # missing data imputation
    ('imputer_num_arbi', mdi.ArbitraryNumberImputer(arbitrary_number=-1, variables=['Cabin_n'])),
    
    ('imputer_num_mean', mdi.MeanMedianImputer(imputation_method='mean', variables=['Age'])),
    
    ('imputer_cat_freq', mdi.CategoricalVariableImputer(imputation_method='missing', variables=['Embarked'])),
    
    ('imputer_cat_missing', mdi.CategoricalVariableImputer(variables=['Cabin_c', 'Title'])),
    
    # Categorical Encoding
    ('encoder_rare_label', RareLabelCategoricalEncoder(tol=0.01, n_categories=4, variables=['Cabin_c', 'Title'])),
    
    ('categorical_encoder', OneHotCategoricalEncoder(top_categories=None, 
                                                     variables=['Sex', 'Embarked','Cabin_c','Title'], drop_last=True)),
    
    ('scaler', SklearnTransformerWrapper(transformer=StandardScaler(), variables=['Fare', 'Age', 'Cabin_n'])),
    
    
])
```

```python
titanic_pipe.fit(X_train, y_train)
```

```python
# Apply tranformations

X_train=titanic_pipe.transform(X_train)
X_test= titanic_pipe.transform(X_test)
```

# Different Models

KNN

```python
from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsClassifier

knn=KNeighborsClassifier()

knn_param={'n_neighbors': range(1,30)}

knn_grid=GridSearchCV(knn, knn_param, cv=6, return_train_score=True)

knn_grid.fit(X_train, y_train)

#Best validation score
print("Best Mean Cross-validation score: {:.2f}%".format(knn_grid.best_score_*100))

#Best parameters
print('KNN parameters: ', knn_grid.best_params_)

print("Accuracy on training set: {:.3f}%".format(knn_grid.score(X_train, y_train)*100))
print("Accuracy on test set: {:.3f}%".format(knn_grid.score(X_test, y_test)*100))

# Check test data set performance
print("KNN Performance: {:.2f}%" .format(knn_grid.score(X_test,y_test)*100))
```

Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
tree=DecisionTreeClassifier()

tree_param={"max_depth": range(1,20),
            "max_leaf_nodes": range(2,20)}

tree_grid= GridSearchCV(tree, tree_param, cv=7, return_train_score= True)

tree_grid.fit(X_train, y_train)

#Best validation score
print("Best Mean Cross-validation score: {:.2f}%".format(tree_grid.best_score_*100))

#Best parameters
print('Decision Tree parameters: ', tree_grid.best_params_)

#Best Score
print("Accuracy on training set: {:.3f}%".format(tree_grid.score(X_train, y_train)*100))
print("Accuracy on test set: {:.3f}%".format(tree_grid.score(X_test, y_test)*100))

# Check test data set performance
print("Decision Tree Performance: {:.2f}%" .format(tree_grid.score(X_test,y_test)*100))
```

SVM

```python
from sklearn.svm import SVC

svc = SVC(kernel = 'rbf')

svc_param = [{'kernel': ['rbf'],'C': [0.001, 0.01, 0.1, 1, 10, 100], 'gamma': [0.001, 0.01, 0.1, 1, 10, 100]}]
             
svc_grid=GridSearchCV(svc,svc_param, cv=6, return_train_score= True)

svc_grid.fit(X_train, y_train)

#Best validation score
print("Best Mean Cross-validation score: {:.2f}%".format(svc_grid.best_score_*100))

#Best parameters
print('SVM parameters: ', svc_grid.best_params_)

#Best Score
print("Accuracy on training set: {:.3f}%".format(svc_grid.score(X_train, y_train)*100))
print("Accuracy on test set: {:.3f}%".format(svc_grid.score(X_test, y_test)*100))

# Check test data set performance
print("SVM Performance: {:.2f}%" .format(svc_grid.score(X_test,y_test)*100))
```
